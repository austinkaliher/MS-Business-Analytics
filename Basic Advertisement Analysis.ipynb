```python
import pandas as pd
import numpy as np
import arrow
import statsmodels.api as sm

tpf = pd.read_csv('travelponyfacebook.csv')

#----------------------------------Question #1---------------------------------

# Create the cost per impression collumn in the tfp dataframe
tpf['Cost per impression'] = tpf['Amount Spent (USD)'] / tpf['Impressions']

# Create a Weekday column on the dataframe so we can analyze cost per impression for different days of the week
   
dayofweek = []

for row in range(len(tpf)):
    date = arrow.get(tpf['Start Date'][row], 'M/D/YY')
    dayofweek.append(date.isoweekday())
    
tpf['Day'] = dayofweek

group = tpf.groupby('Day')
print(group.aggregate(np.mean))

print(' ')
print('cost per impression is minimzed on day 6 which is saturday in the isoweekday format.')
print('cost per impression is maximied on day 5 which is friday in the isoweekday format.')

print(' ')
print(' ')
print(' ')
print(' ')

#---------------------------------Question #2----------------------------------
# correlation between ammount spent and reach

amountreachcorr = np.corrcoef(tpf['Amount Spent (USD)'], tpf['Reach'])
print('The correlation coefficient between Amount Spent and Reach is:')
print(amountreachcorr)

# correlation between amount spent and frequency

amountfrequencycorr = np.corrcoef(tpf['Amount Spent (USD)'], tpf['Frequency'])
print('The correlation coefficient between Amount Sepnt and Frequency is:')
print(amountfrequencycorr)

# correlation between amount spent and unique clicks

amountuniqueclickscorr = np.corrcoef(tpf['Amount Spent (USD)'], tpf['Unique Clicks'])
print('The correlation coefficient between Amount Spent and Unique Clicks is:')
print(amountuniqueclickscorr)

# correlation between amount spent and page likes

amountpagelikescorr = np.corrcoef(tpf['Amount Spent (USD)'], tpf['Page Likes'])
print('The correlation coefficient between Amount Spent and Page Likes is:')
print(amountpagelikescorr)
print(' ')
print(' ')
print(' ')
print(' ')
print(' ')

#---------------------------------Question #3----------------------------------

y = tpf['Unique Clicks']
x = tpf[['Reach', 'Frequency']]

model = sm.OLS(y,x).fit()
predictions = model.predict(x)

print(model.summary())


print(' ')
print(' ')
print(' ')
print(' ')
print('Frequency most strongly predicts unique clicks because the coefficient for Frequency is much larger than reach')
```
