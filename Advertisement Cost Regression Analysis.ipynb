### This is a python script originally created for APRD 6342 (Digital Advertising) at the University of Colorado Boulder. In this assignment, we were tasked with finding out what drives clicks on Facebook advertisements for an online travel booking company. To do this, we constructed a multiple linear regression. The data for this code is available in my repository. Included below are the original questions and answers.

```python
import pandas as pd
import statsmodels.api as sm
import seaborn as sns


tpf = pd.read_csv('Travel Pony Facebook.csv')

#
#
#
#--------------------------------Question #1-----------------------------------
#
#
#

print(' ')
print(' ')
print(' ')
print('========================Question #1, Histogram========================')
print(' ')
print(' ')
print(' ')

# remove the rows that have $0 spent
tpf = tpf[ tpf['Amount Spent (USD)'] != 0 ]
tpf['amtspent'] = tpf['Amount Spent (USD)']

#1. Next, graph a histogram of Amount Spent using Seaborn to visualize the
#   typical amount spent on Facebook ads
print('This is a plot of amount spent in USD for the ad campaigns in the Travel') 
print('Pony facebook data set. If you don\'t see the graph here, please see')
print('the bottom of the output')
print(' ')
print(' ')
print(' ')
sns.set_style('darkgrid')
print(sns.distplot(tpf['amtspent']))

#
#
#
#--------------------------------Question #2-----------------------------------
#
#
#

# Build a multiple regression where the outcome variable is Amount Spent 
# and the predictor variables are:

# Campaign Name, Reach, Frequency, Impressions, Clicks, Unique Clicks, 
# Page Likes, Page Engagement, Post Engagement, Post Likes, Post Comments, 
# Post Shares, Photo Views, Website Clicks

# Make sure that you convert "Campaign Name" to some type of categorical value
#that the regression understands as such

print(' ')
print(' ')
print(' ')
print('================Question #2, Multiple Linear Regression================')
print(' ')
print(' ')
print(' ')

tpf['Campaign Name'] = tpf['Campaign Name'].astype('category')
tpf['Campaign_Code'] = tpf['Campaign Name'].cat.codes
x = tpf[['Campaign_Code', 'Reach', 'Frequency', 'Impressions', 'Clicks'
         , 'Unique Clicks', 'Page Likes', 'Page Engagement', 'Post Engagement'
         , 'Post Likes', 'Post Comments', 'Post Shares', 'Photo Views'
         , 'Website Clicks']]
y = tpf['amtspent']
x = sm.add_constant(x)
x.shape
linreg=sm.OLS(y,x).fit()

print(linreg.summary())

#
#
#
#--------------------------------Question #3-----------------------------------
#
#
#

# Look at the coefficients for each one of your predictor variables. 
# The higher the value, the stronger predictive value the variable has with 
# amount spent. That is, when we spend more, we tend to get more of this variable
# than anything else. What are the three predictors with the highest predictive 
# value (rank them, please)

print(' ')
print(' ')
print(' ')
print('===============================Question #3===============================')
print(' ')
print(' ')
print(' ')
print('When taking into consideration the magnitude of the coefficients, the')
print('top three predictors are page likes, page engagement, and post engagement')
print(' ')
print(' ')
print(' ')
print('   Predictor             Coefficient')
print(' ')
print('   Page Likes              23.0417')
print(' Page Engagement          -19.7454')
print(' Post Engagement           16.8084')
print(' ')
print(' ')
print(' ')
print('=========================================================================')
print(' ')
print(' ')
print(' ')
print('One could argue that a significant negative coefficient for Page')
print('Engagement doesn\'t make sense given the context of the question.') 
print('If we were to remove page engagement from the list, the top three') 
print('predictors are page likes, post engagement, and frequency. These are')
print('top three variables with positive coefficients. This is technically')
print('the answer to the prompt.')
print(' ')
print(' ')
print(' ')
print('   Predictor             Coefficient')
print(' ')
print('   Page Likes              23.0417')
print(' Post Engagement           16.8084')
print('   Frequency                4.4988')
print(' ')
print(' ')
print(' ')
print('=========================================================================')
```
